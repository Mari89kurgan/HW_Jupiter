### 1.1. Данные из прошлого дз

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv', sep=',')
df
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(5, 5))
plt.hist(df['price'])
plt.title('Распределение цен на недвижимость')
plt.xlabel('Цена')
plt.ylabel('Количество');
```

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(7, 9))
sns.histplot(df['sqft_living'], bins=20)
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('Жилая площадь, в кв.м.')
plt.ylabel('Количество');
```

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(7, 9))
sns.histplot(df['yr_built'], bins=30)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество');
```

## Условие 2: 2 задача

### 2.1 Изучите распределение домов от наличия вида на набережную

```python
data = df['waterfront'].value_counts()

plt.figure(figsize=(4, 4))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['no', 'yes']);
```

Очень маленький процент (0,8%) от всех домов имеет вид на набережную.

### 2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(data, autopct='%1.1f%%', labels=data.index)
plt.title('Количество этажей');
```

### 2.3 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%', labels=data.index);
```

## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Чем больше жилая площадь, тем выше стоимость недвижимости.

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Чем выше качество конструкции и дизайна дома, тем выше цена.
